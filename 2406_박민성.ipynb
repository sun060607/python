```python
filename = "/content/drive/MyDrive/age.csv"
import csv
f = open(filename, 'r',encoding='cp949')
data = csv.reader(f)
head = next(data)
result1=[]
result2=[]
result3=[]
result4=[]
result5=[]

loc_name1 = '서울특별시 종로구 혜화동'
loc_name2 = '서울특별시 동대문구 이문제1동'
loc_name3 = '서울특별시 동대문구 회기동'
loc_name4 = '서울특별시 중구 장충동'
loc_name5 = '서울특별시 성북구 동선동'
for row in data:
  if loc_name2 in row[0]:
    for i in row[3:]:
      result2.append(int(i.replace(',','')))
  if loc_name3 in row[0]:
    for i in row[3:]:
      result3.append(int(i.replace(',','')))
  if loc_name4 in row[0]:
    for i in row[3:]:
      result4.append(int(i.replace(',','')))
  if loc_name5 in row[0]:
    for i in row[3:]:
      result5.append(int(i.replace(',','')))
  if loc_name1 in row[0]:
    for i in row[3:]:
      result1.append(int(i.replace(',','')))

f.close()
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.plot(result1,label='서울특별시 종로구 혜화동(1111065000)')
plt.plot(result2,label='서울특별시 동대문구 이문제1동(1123074000)')
plt.plot(result3,label='서울특별시 동대문구 회기동(1123071000)')
plt.plot(result4,label='서울특별시 중구 장충동(1114058000)')
plt.plot(result5,label='서울특별시 성북구 동선동(1129057500)')
plt.legend()
plt.show()
```

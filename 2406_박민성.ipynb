```python
filename = "/content/drive/MyDrive/age.csv"
import csv
f = open(filename, 'r',encoding='cp949')
data = csv.reader(f)
head = next(data)
result1=[]
result2=[]
result3=[]
result4=[]
result5=[]

loc_name1 = '  '
loc_name2 = '  1'
loc_name3 = '  '
loc_name4 = '  '
loc_name5 = '  '
for row in data:
  if loc_name2 in row[0]:
    for i in row[3:]:
      result2.append(int(i.replace(',','')))
  if loc_name3 in row[0]:
    for i in row[3:]:
      result3.append(int(i.replace(',','')))
  if loc_name4 in row[0]:
    for i in row[3:]:
      result4.append(int(i.replace(',','')))
  if loc_name5 in row[0]:
    for i in row[3:]:
      result5.append(int(i.replace(',','')))
  if loc_name1 in row[0]:
    for i in row[3:]:
      result1.append(int(i.replace(',','')))

f.close()
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.plot(result1,label='  (1111065000)')
plt.plot(result2,label='  1(1123074000)')
plt.plot(result3,label='  (1123071000)')
plt.plot(result4,label='  (1114058000)')
plt.plot(result5,label='  (1129057500)')
plt.legend()
plt.show()
```

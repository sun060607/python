- 리스트 생성 : []또는 리스트 함수 이용
- 인덱싱
 - 요소의 선택은 0부터 시작되는 인덱스로 접근
 - 리스트명[인덱스]
- 슬라이싱
 - 요소의 일부분 선택
 - 리스트명[시작인덱스:종료인덱스]
 - 종료 인덱스는 포함되지 않음
- 연산
 - +연산:리스트와 리스트 연결
 - *연산:리스트 요소 반복
- 추가
 - 리스트명.append(추가요소):마지막에 추가
 - 리스트명.insert(인덱스,추가요소):해당 인덱스에 요소 추가
- 삭제
 - 리스트명.pop():마지막 요소 삭제
 - 리스트명.pop(인덱스):해당 인덱스 요소 삭제
 - 리스트명.remove(요소값):해당 요소값 삭제, 해당 값이 없으면 오류
- 수정
 - 리스트명[인덱스] = 변경값
- 위치 반환
 - 리스트명.index(요소값)
 - 해당 요소값의 위치를 반환
 - 여러 개 요소가 존재하면 첫 번째 위치 반환하고 없으면 오류
- 요소 개수 구하기
 - 리스트명.count(요소값)
- 정렬
 - 리스트명.sort():오름차순
 - 리스트명.sort(reverse=True):내림차순
- 뒤집기
 - 리스트명.reverse():요소를 역순으로 뒤집어 줌

```python
#리스트 추가
#리스트.append(추가 요소) : 마지막에 추가
#리스트명.insert(인덱스,. 추가요소) : 해당 인덱스에 요소 추가
my_list1=['a','b','c']
my_list1.append('d')
print(my_list1)
my_list1.insert(0,'z')
print(my_list1)

my_time=[1,2,3]
my_temp=[20,20,25]
my_humi=[60,70,80]
```

```python
#리스트의 삭제
#리스트명.pop() : 마지막 요소 삭제
#리스트명.pop()인덱스 : 해당 인덱스 요소 삭제
#리스트명.remove(요소값) : 해당 요소값 삭제, 해당 값이 없으면 오류
my_list1=['a','b','c']
my_list1.pop()
print(my_list1)
my_list1.pop(0)
print(my_list1)
my_list1.remove('b')
print(my_list1)
```

```python
#리스트 수정
#리스틈명[인덱스] = 변경값
list = ['a','b','c']
list[0] = 'e'
print(list)
```

```python
#리스트 정렬
#리스트.sort() : 오름차순
#리스트.sort(reverse=True) :내림차순
list = ['a','b','c']
list.sort()
print(list)
list.sort(reverse=True)
print(list)
```

```python
#리스트 뒤집기
#리스트명reverse() : 요소를 역순으로 뒤집어줌
list = ['a','b','c']
list.reverse()
print(list)
print(list[-1::-1])
```

#3) 딕셔너리 


*   특징
    + 키(key)와 값(value) 쌍을 요소로 가짐
    + 중괄호로 작성되며 각 요소는 쉼표로 구분
    + 순서 없음
    + 키는 변경할 수 없으며 값은 변경가능
*   딕셔너리 접근
    + 특정 요소 값 : 딕셔너리명[키]
    + 요소 키 컬렉션 : 딕셔너리명.keys()
    + 요소 값 컬렉션 : 딕셔너리.values()
    + 키와 요소 컬렉션은 list() 생성자로 리스트로 변환 가능
*   요소 추가
    + 딕셔너리명[추가키] = 값
*   요소 수정
    + 딕셔너리명[키] = 값
*   요소 삭제
    + del딕셔너리명[키]
    + 딕셔너리명.pop(키)



```python
#딕셔너리
dt={'a':200,'b':150,'c':100}
print(dt)
```

```python
#딕셔너리 키
dt={'a':200,'b':150,'c':100}
#print(list(dt.values()))
```

```python
# 딕셔너리 접근
dt={'a':200,'b':150,'c':100}
print(dt['a'])
print(dt['b'])
print(dt['c'])
```

```python
#딕셔너리 추가
dt={'a':200,'b':150,'c':100}
dt['d'] = 300
print(dt)
dt['b'] = 400
print(dt)
dt.update({'e':500})
print(dt)
```

```python
#딕셔너리 삭제
dt={'a':200,'b':150,'c':100}
del dt['b']
print(dt)
dt.pop('a')
print(dt)
#모두 지우기
dt.clear()
print(dt)
```

#4) 집합



*   특징
    + 중복 없는 요소로만 구성
    + 중괄호로 작성되며 각 요소는 쉼표로 구분
    + 순서 없음
*   생성자 : set()
    + 리스트나 튜플의 중복 요소 제거
*   요소 추가
    + 집합명.add(요소) : 하나의 요소 추가
    + 집합명.update(추가집합) : 여러 요소를 추가
*   요소 삭제
    + 집합명.remove(요소값) : 하나의 요소 삭제
    + 집합명.clear() : 모든 요소 삭제
*   집합 연산
    + 교집합: &
    + 합집합: |
    + 차집합: -



```python
#집합
a=(1,2,2,3)  #튜플
b=[10,20,30,30]  #리스트
a=set(a)
b=set(b)

#집합 요소 추가
print(a)
print(b)
a.add(100)
b.update({100,200,300})
print(a)
print(b)

#집합 요소 삭제
a.remove(100)
print(a)
b.clean()
print(b)
```

```python
#집합 연산
a=(1,2,2,3)  #튜플
b=[10,20,30,30]  #리스트
a=set(a)
b=set(b)

print(f'a집합={a}, b집합={b}')
print(f'교집합 a & b = {a & b}')
print(f'교집합 a | b = {a | b}')
print(f'교집합 a - b = {a - b}')
print(f'교집합 b - a = {b - a}')
```

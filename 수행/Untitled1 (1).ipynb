```python
filename = "/content/drive/MyDrive/gender.csv"
import csv
f = open(filename, 'r',encoding='cp949')
data = csv.reader(f)
head = next(data)
loc_name = "부산광역시 연제구 연산제1동"

sum = []
for row in data:
  if loc_name in row[0]:
    male = 0
    female = 0
    for i in range(0,101):
      male += int(row[i+3].replace(',',','))
      female += int(row[i+106].replace(',',','))
sum.append(male)
sum.append(female)
f.close()

import matplotlib.pyplot as plt
plt.rc('font', family="NanumBarunGothic")
plt.rcParams['axes.unicode_minus'] = False

label = ['남성','여성']
color = ['skyblue','pink']
plt.axis('equal')
plt.pie(sum,labels=label,colors = color,autopct='%.1f%%',startangle=90,explode=(0,0.1))
plt.legend()
plt.show()
```
